Iako su svi atributi zapisani u obliku float64, vidimo da su zapravo tipa int64 ili object.

```python
# Kategoricki atributi
categorical = [
    'Diabetes_012', 
    'HighBP', 
    'HighChol', 
    'CholCheck', 
    'Smoker', 
    'Stroke', 
    'HeartDiseaseorAttack', 
    'PhysActivity', 
    'Fruits', 
    'Veggies', 
    'HvyAlcoholConsump', 
    'AnyHealthcare', 
    'NoDocbcCost', 
    'GenHlth',
    'DiffWalk',
    'Sex',
    'Age',
    'Education',
    'Income'
]
```

```python
data[['BMI', 'PhysHlth', 'MentHlth']] = data[['BMI', 'PhysHlth', 'MentHlth']].astype('int64')
data[categorical] = data[categorical].astype('int64').astype(object)
```

```python
data.info()
```

# Podela numeričkih atributa u kategorije

```python
for c in data.columns:
    print(c, end=" ")
    print(data[c].unique())
```

## BMI

Na osnovu domenskog znanja, možemo da pretpostavimo da nam za klasifikaciju neće biti neophodno da znamo tačan BMI, već samo kojoj kategoriji pripada:
- 0: BMI <= 18
- 1: 19 <= BMI <= 24
- 2: 25 <= BMI <= 29
- 3: 30 <= BMI <= 39
- 4: BMI > 40

```python
data['BMI'] = pd.cut(x = data['BMI'], bins = [0, 18, 24, 29, 39, 100], labels = [0, 1, 2, 3, 4])
```

```python
data['BMI']
```

## PhysHlth i MentHlth

Ovi atributi predstavljaju broj dana u prethodnih 30 za koje važi da je osoba imala problema sa fizičkim/mentalnim zdravljem. Možemo da pokušamo da radimo sa podacima u ovom obliku, ali možemo da pokušamo i da ih podelimo u neke kategorije.

Kada je u pitanju mentalno zdravlje i postavljanje dijagnoza, često se kao kriterijum uzima da su simptomi prisutni oko 2 nedelje. Napravicemo podelu tako da sve do 10 dana tretiramo kao normalno, a da preko 10 dana tretiramo kao problem. Slično ćemo uraditi i za fizičko zdravlje.

```python
data['PhysHlth'] = pd.cut(x = data['PhysHlth'], bins = [-1, 15, 31], labels = [0, 1])
data['MentHlth'] = pd.cut(x = data['MentHlth'], bins = [-1, 15, 31], labels = [0, 1])
```
