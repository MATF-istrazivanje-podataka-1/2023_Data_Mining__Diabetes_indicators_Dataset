```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('../0. Data/data_balanced.csv')
```

```python
data = data.drop('Unnamed: 0', axis=1)
```

```python
data
```

```python
scaler = StandardScaler()
```

```python
data_standardized = scaler.fit_transform(data)
```

```python
data_standardized = pd.DataFrame(data=data_standardized, columns=data.columns)
```

```python
data_standardized
```
